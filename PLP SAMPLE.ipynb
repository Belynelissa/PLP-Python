## Import required libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## Load the dataset

```python
# Load the dataset
df = pd.read_csv("data.csv")

# Display first few rows
print("First 5 rows of the dataset:")
print(df.head())
```

## Exploratory data analysis

```python
# Display dataset info
print("\n🔹 Dataset Information:")
print(df.info())

# Check for missing values
print("\n🔹 Missing Values:")
print(df.isnull().sum())


# Display summary statistics
print("\n🔹 Summary Statistics:")
print(df.describe())
```

## Basic statistical analysis

```python
# Compute average CO2 emissions based on car model
model_avg_co2 = df.groupby("Model")["CO2"].mean()
print("\n🔹 Average CO2 Emissions per Car Model:")
print(model_avg_co2)
```

## Data visualisation

```python
print("🔹 Available Columns in Dataset:")
print(df.columns)
```

### Line Chart (CO2 Emissions Trend by Car Volume

```python
plt.figure(figsize=(10, 5))
sns.lineplot(x="Volume", y="CO2", data=df, marker="o")
plt.title("📈 CO2 Emissions vs. Car Volume")
plt.xlabel("Car Volume")
plt.ylabel("CO2 Emissions")
plt.show()
```

### Bar Chart (Average CO2 Emissions by Model)

```python
plt.figure(figsize=(10, 5))
sns.barplot(x=model_avg_co2.index, y=model_avg_co2.values, palette="coolwarm")
plt.title("📊 Average CO2 Emissions by Car Model")
plt.xlabel("Car Model")
plt.ylabel("CO2 Emissions")
plt.xticks(rotation=45)
plt.show()
```

### Histogram (Distribution of Car Weights)

```python
plt.figure(figsize=(8, 5))
sns.histplot(df["Weight"], bins=20, kde=True, color="green")
plt.title("📊 Car Weight Distribution")
plt.xlabel("Weight")
plt.ylabel("Frequency")
plt.show()
```

###  Scatter Plot (Relationship Between Car Weight & CO2 Emissions)

```python
plt.figure(figsize=(8, 5))
sns.scatterplot(x="Weight", y="CO2", data=df, alpha=0.7)
plt.title("🔍 Relationship Between Car Weight and CO2 Emissions")
plt.xlabel("Car Weight")
plt.ylabel("CO2 Emissions")
plt.show()
```
